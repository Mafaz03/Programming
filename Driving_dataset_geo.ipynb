```python
from scipy.stats import chi2_contingency
import numpy as np

# Create a contingency table
contingency_table = pd.crosstab(com_data['Alert'], com_data['road'])

# Calculate the chi-squared statistic and p-value
chi2, _, _, _ = chi2_contingency(contingency_table)

# Calculate Cramer's V
n = contingency_table.sum().sum()  # Total number of observations
min_dim = min(contingency_table.shape) - 1  # Minimum of rows and columns in the contingency table
cramers_v = np.sqrt(chi2 / (n * min_dim))

print(f"Cramer's V: {cramers_v}")
```

```python
def add_highway_column(df, column_name, words_to_check, new_column_name ):
  """
  Adds a new column to a DataFrame based on the presence of words in a specified column.

  Args:
  df (pd.DataFrame): The DataFrame to modify.
  column_name (str): The name of the column to check for the words.
  words_to_check (list): A list of words to check for in the specified column.
  new_column_name (str): The name of the new column to be added.

  Returns:
  pd.DataFrame: The modified DataFrame with the new column.
  """
  def check_words_in_column(row):
    column_value = str(row[column_name])
    for word in words_to_check:
      if word in column_value:
        return 1
    return 0
  df[new_column_name] = df.apply(check_words_in_column, axis=1)
  return df


```
